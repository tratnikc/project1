**Preferred Time of Day to Bike (per quarter)**

```python
bins = [0,2,5,8,11,14,17,20,23]

# create labels for the bins
midnight = "00:00 - 02:59"
three_am = "03:00 - 05:59"
six_am = "06:00 - 08:59"
nine_am = "09:00 - 11:59"
noon = "12:00 - 14:59"
three_pm = "15:00 - 17:59"
six_pm = "18:00 - 20:59"
nine_pm = "21:00 - 23:59"

time_bins =[midnight, three_am, six_am, nine_am, noon, three_pm, six_pm, nine_pm]
```

**Q1**

```python
# Q: What is the peak time for biking in Q1

# convert column data into datetime, then assign to a variable
q1_start_time_df = q1[["Departure"]]
q1_start_time = pd.to_datetime(q1["Departure"])

# get the hour portion and create a data frame
q1_time_started = q1_start_time.dt.hour
q1_time_started_df = pd.DataFrame(q1_time_started)

# merge hour data to start_time_df
q1_rides_df = q1_start_time_df.merge(q1_time_started_df,
                                     left_index=True, right_index=True)

q1_rides_df = q1_rides_df.rename(columns={"Departure_x" : "Start Time",
                                             "Departure_y" : "Hour"})

#display ride_times_df
q1_rides_df

#get total number of rides for Q1
q1_total_rides = len(q1_rides_df)
q1_total_rides

# slice the data and place it into bins
q1_rides_df["Time Group"] = pd.cut(q1_rides_df["Hour"], bins, labels=time_bins, include_lowest=True)

q1_rides = q1_rides_df["Time Group"].value_counts()
q1_rides = q1_rides.sort_index()

#get the percentage of riders per time group
q1_pcts = (q1_rides / q1_total_rides * 100).map("{:.2f}%".format)

# create summary data frame to hold results
q1_hour_bin_df = pd.DataFrame({"Total Rides" : q1_rides, "Percentage of Rides" : q1_pcts})

# display rides summary table
q1_hour_bin_df
```

```python
#Plotting the most popular time of day in Q1

distribution = q1_hour_bin_df["Total Rides"]
time_labels = ["00:00 - 02:59","03:00 - 05:59","06:00 - 08:59",
               "09:00 - 11:59","12:00 - 14:59", "15:00 - 17:59","18:00 - 20:59","21:00 - 23:59"] 
explode = (0, 0.20, 0, 0,0,0,0,0)
colors = ("#FFF1C9", "#EA5F89","#F7B7A3","#AADEA7","#2D87BB","#7AC142","#FDBB2F", "#F47A1F")
fig, ax = plt.subplots(figsize=(10,6))
ax.pie(distribution, labels=time_labels,explode=explode,wedgeprops=dict(width=0.6),colors = colors,
       textprops=dict(color="black",size=12),labeldistance = 1.1, pctdistance =0.8,counterclock=True,autopct="%1.1f%%")
plt.axis("equal")
plt.title("MOST POPULAR TIME OF DAY IN Q1", fontsize=14, weight="bold")


#Save the file as .png format
file = os.path.join("Images","Most popular time of day in Q1 Vancouver.png")
plt.savefig(file)

plt.show()
```

```python
#Convert the objects into DateTime format:
#bike_share.drop(bike_share["Return"]=="1900-01-00 0:00")
bike_share['Departure'] = pd.to_datetime(bike_share['Departure'])
bike_share['Return']= pd.to_datetime(bike_share['Return'])
```

```python
#extract the hour, month and day of the week from both columns:
bike_share['hour_start'] = bike_share['Departure'].apply(lambda time: time.hour)
bike_share['month_start'] = bike_share['Departure'].apply(lambda time: time.month)
bike_share['weekday_start'] = bike_share['Departure'].apply(lambda time: time.dayofweek)
```

```python
bike_share['hour_stop'] = bike_share['Departure'].apply(lambda time: time.hour)
bike_share['month_stop'] = bike_share['Departure'].apply(lambda time: time.month)
bike_share['weekday_stop'] = bike_share['Departure'].apply(lambda time: time.dayofweek)
```

```python
mon = {1:'Jan',2:'Feb',3:'Mar',4:'Apr',5:'May',6:'Jun',7:'Jul',8:'Aug',9:'Sep',10:'Oct',11:'Nov',12:'Dec'}
bike_share['month_start'] = bike_share['month_start'].map(mon)
bike_share['month_stop'] = bike_share['month_stop'].map(mon)
```

```python
day = {0:'Mon',1:'Tue',2:'Wed',3:'Thu',4:'Fri',5:'Sat',6:'Sun'}
bike_share['weekday_start'] = bike_share['weekday_start'].map(day)
bike_share['weekday_stop'] = bike_share['weekday_stop'].map(day)
```
